```python
def distEuc(li1,li2):
    res=0
    for i in range(len(li2)):
        res+=pow(li1[i]-li2[i],2)
    return sqrt(res)
```

```python
def distance(liste,df):
    res=[]
    for i in range(len(df)): 
        li_attr=[]
        for j in range (0,4):
            li_attr.append(df.iloc[i,j]) 
            res.append(distEuc(liste,li_attr))
    return res     
```
